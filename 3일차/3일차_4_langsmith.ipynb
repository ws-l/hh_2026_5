## 1. Langsmith

- LangSmith 개요: LangSmith = LLM용 로그 + 디버거 + 테스트 도구 를 하나로 합친 것
  
    - LLM 애플리케이션의 디버깅 · 모니터링 · 평가 플랫폼
    - LangChain이 만든 LLMOps 도구
    - 코드 변경 없이 환경변수 설정만으로 활성화

- Tracing (트레이싱)
    - LLM 호출 전체 흐름을 자동 기록
    - 입력 · 출력 · 소요시간 · 토큰 수 · 비용 추적
    - 노드별 실행 결과 시각화 (LangGraph와 잘 호환)

- Playground
    - 기록된 트레이스를 클릭 → 즉시 재실행
    - 프롬프트 수정 후 결과 비교 가능
    - 코드 없이 브라우저에서 실험

- Datasets & Evaluation
    - 테스트 데이터셋 구축 · 관리
    - 자동 평가 파이프라인 구성
    - 정확도 · 품질 지표 추적

- Monitoring
    - 프로덕션 환경 실시간 모니터링
    - 오류율 · 지연시간 · 토큰 사용량 대시보드
    - 알림 설정 가능
