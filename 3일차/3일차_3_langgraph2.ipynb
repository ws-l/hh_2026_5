# 3. Langgraph 개요


- LLM 애플리케이션을 그래프 구조로 설계하는 프레임워크
    - 노드(Node) = 작업 단위
    - 엣지(Edge) = 흐름 연결
    - LangChain 위에서 동작, 상태(State)를 노드 간에 공유  
      
        
      
- 핵심 구성요소
    - State: 노드 간에 공유되는 데이터 저장소, 대화 맥락을 전달하는 공유메모리
        - TypedDict로 정의
        - 각 노드는 State를 받아서 일부만 업데이트해서 반환

        - 예

````python
# 시작 시 초기 State
{"question": "CNC Burr 원인은?", "route": "", "context": "", "answer": ""}

# classify 노드 실행 후
{"question": "CNC Burr 원인은?", "route": "rag", "context": "", "answer": ""}
#                                              ↑ 이것만 업데이트

# rag 노드 실행 후
{"question": "CNC Burr 원인은?", "route": "rag", "context": "검색결과...", "answer": ""}
#                                                              ↑ 이것만 업데이트

# answer 노드 실행 후
{"question": "CNC Burr 원인은?", "route": "rag", "context": "검색결과...", "answer": "Burr는..."}
#                                                                             ↑ 이것만 업데이트
````

- (이어서)
    - Node
        - 실제 작업을 수행하는 함수
        - State를 입력받아 → 처리 후 → State 일부 반환
        - LLM 호출, DB 조회, 검색 등 모든 작업이 노드로 표현

    - Edge
        - 노드 간 연결 (실행 순서 정의)
        - 일반 엣지: 항상 같은 다음 노드로 이동
        - 조건부 엣지: 상태값에 따라 분기

- Agent와 Langgraph와의 차이
    - Agent는 흐름을 LLM이 결정, 분기가 어렵고 루프 제한적임
    - Langgraph는 개발 시 명시적 설계, 그래프로 시각화, 조건부 엣지로 분기 가능, 노드별 state 추적으로 디버깅

- 주요 패턴
    - 분류 → 라우팅 패턴
        - 질문 → classify → rag / sql / ml → answer
    - ReAct 패턴 (create_react_agent): Reasoning + Acting, LLM이 스스로 생각 → 도구선택 → 실행 → 관찰 을 반복하며 답을 찾는 패턴
        - 질문 → LLM → tool 선택 → tool 실행 → LLM → 반복 → 최종답변

    - 체인 패턴
            - 노드A → 노드B → 노드C → 순차 실행
    - 루프 패턴
            - 생성 → 평가 → 품질 미달이면 생성으로 재시도 → 통과하면 종료

- 실행 흐름
    - builder = StateGraph(AgentState)   # 그래프 생성
    - builder.add_node(...)              # 노드 등록
    - builder.add_edge(...)              # 엣지 연결
    - graph = builder.compile()          # 컴파일
    - graph.invoke({"question": "..."})  # 실행

- 언제 사용?
    - Agent 흐름을 명시적으로 제어하고 싶을 때
    - 질문 유형에 따라 다른 도구/경로로 분기할 때
    - 생성 → 검증 → 재시도 같은 루프가 필요할 때
    - 복잡한 멀티 에이전트 파이프라인을 구성할 때


- 구조  

START  
  ↓  
classify  (LLM이 질문 의도 분류)  
  ↓  
  ├─ rag    → lightrag_search() → answer  
  ├─ sql    → sql_search()      → answer    
  └─ python → python_analysis() → answer  
                                    ↓  
                                    END

````python
from typing import TypedDict, Literal
import sqlite3
import pandas as pd

from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama


llm = ChatOllama(    model="qwen2.5:3b",    temperature=0)
embeddings = OllamaEmbeddings(model="nomic-embed-text")
````

````python
# 벡터스토어 생성
vectorstore = Chroma(
    collection_name="factory_docs",
    embedding_function=embeddings,
    persist_directory="./chroma_db1"
)
````

````python
# reports.json 로드
with open("reports.json", "r", encoding="utf-8") as f:
    reports = json.load(f)

documents = [
    Document(
        page_content=r["report_text"],
        metadata={"report_id": r["report_id"], "source": "json"}
    )
    for r in reports
]
vectorstore.add_documents(documents)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

print(f"ChromaDB 구성 완료: {vectorstore._collection.count()}개")
````

- duckdb 구성

````python
df_logs = pd.read_csv("work_logs.csv", encoding="utf-8-sig")

con = duckdb.connect("factory.db")
con.execute("DROP TABLE IF EXISTS work_logs")
con.execute("CREATE TABLE work_logs AS SELECT * FROM df_logs")
print(f"DuckDB 구성 완료: {con.execute('SELECT COUNT(*) FROM work_logs').fetchone()[0]}건")
con.close()
````

- ML모형 학습

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

con = duckdb.connect("factory.db", read_only=True)
df = con.execute("SELECT * FROM work_logs").df()
con.close()

le_설비 = LabelEncoder()
le_작업 = LabelEncoder()
df["설비명_enc"] = le_설비.fit_transform(df["설비명"])
df["작업유형_enc"] = le_작업.fit_transform(df["작업유형"])
df["불량"] = (df["작업결과"] != "완료").astype(int)

X = df[["설비명_enc", "작업유형_enc", "소요시간_h", "비용_원"]]
y = df["불량"]

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

with open("defect_model.pkl", "wb") as f:
    pickle.dump({"model": model, "le_설비": le_설비, "le_작업": le_작업}, f)

print(f"ML 모델 학습 완료 | 불량 비율: {y.mean():.1%}")
````

- State 정의

````python
class AgentState(TypedDict):
    question: str
    route: str
    context: str
    answer: str
````

- 노드함수 정의

````python
# ── RAG 노드 ──────────────────────────────
def rag_node(state: AgentState):
    docs = retriever.invoke(state["question"])
    context = "\n".join([d.page_content for d in docs])
    return {"context": context}


# ── SQL 노드 ──────────────────────────────
def sql_node(state: AgentState):
    try:
        con = duckdb.connect("factory.db", read_only=True)
        # LLM에게 SQL 생성 요청
        sql_prompt = f"""
        아래 테이블에서 질문에 답하는 SQL을 작성하세요.
        테이블: work_logs
        컬럼: log_id, 작업일시, 설비명, 작업유형, 작업자, 소요시간_h, 작업결과, 비용_원

        질문: {state["question"]}

        SELECT 문만 출력하세요. 설명 없이.
        """
        sql = llm.invoke(sql_prompt).content.strip()
        # 코드블록 제거
        sql = sql.replace("```sql", "").replace("```", "").strip()

        result = con.execute(sql).df()
        con.close()
        return {"context": result.to_string(index=False)}
    except Exception as e:
        return {"context": f"SQL 오류: {str(e)}"}


# ── ML 예측 노드 ───────────────────────────
def ml_node(state: AgentState):
    try:
        with open("defect_model.pkl", "rb") as f:
            artifacts = pickle.load(f)
        clf = artifacts["model"]
        le_설비 = artifacts["le_설비"]
        le_작업 = artifacts["le_작업"]

        con = duckdb.connect("factory.db", read_only=True)
        df_recent = con.execute("""
            SELECT 설비명, 작업유형, 소요시간_h, 비용_원
            FROM work_logs
            ORDER BY 작업일시 DESC
            LIMIT 3
        """).df()
        con.close()

        results = []
        for _, row in df_recent.iterrows():
            설비_enc = le_설비.transform([row["설비명"]])[0]
            작업_enc = le_작업.transform([row["작업유형"]])[0]
            X = [[설비_enc, 작업_enc, row["소요시간_h"], row["비용_원"]]]
            prob = clf.predict_proba(X)[0][1]
            results.append(f"{row['설비명']} / {row['작업유형']}: 불량확률 {prob:.1%}")

        return {"context": "\n".join(results)}
    except Exception as e:
        return {"context": f"예측 오류: {str(e)}"}

````

````python
# ── 분류 노드 ──────────────────────────────
def classify_intent(state: AgentState):
    prompt = f"""
    다음 질문을 세 가지 중 하나로만 분류하세요.

    - rag: 불량 원인, 조치방법, 보고서 내용 검색
    - sql: 설비별 작업이력, 건수 집계, DB 조회
    - ml: 불량 예측, 위험도 분석

    질문: {state["question"]}

    rag, sql, ml 중 하나만 출력하세요.
    """
    response = llm.invoke(prompt).content.strip().lower()

    if "sql" in response:
        route = "sql"
    elif "ml" in response:
        route = "ml"
    else:
        route = "rag"

    print(f"라우팅: {route}")
    return {"route": route}


# ── 최종 답변 노드 ─────────────────────────
def answer_node(state: AgentState):
    prompt = f"""
    사용자의 질문에 아래 정보를 참고하여 한국어로 답하세요.

    질문: {state["question"]}
    참고 정보: {state["context"]}

    간결하고 명확하게 답변하세요.
    """
    response = llm.invoke(prompt)
    return {"answer": response.content}


# ── 라우터 ─────────────────────────────────
def router(state: AgentState) -> Literal["rag", "sql", "ml"]:
    return state["route"]
````

- 그래프 구성

````python
builder = StateGraph(AgentState)

builder.add_node("classify", classify_intent)
builder.add_node("rag",      rag_node)
builder.add_node("sql",      sql_node)
builder.add_node("ml",       ml_node)
builder.add_node("answer",   answer_node)

builder.add_edge(START, "classify")
builder.add_conditional_edges(
    "classify",
    router,
    {"rag": "rag", "sql": "sql", "ml": "ml"}
)
builder.add_edge("rag",    "answer")
builder.add_edge("sql",    "answer")
builder.add_edge("ml",     "answer")
builder.add_edge("answer", END)

graph = builder.compile()
print("그래프 컴파일 완료")
````

- 실행

````python
questions = [
    "CNC에서 Burr가 발생하는 원인은?",          # → rag
    #"프레스 P-2 작업이력 건수 알려줘",            # → sql
    #"최근 작업 데이터로 불량 위험도 예측해줘",     # → ml
]

for q in questions:
    print(f"\n{'='*50}")
    print(f"질문: {q}")
    result = graph.invoke({"question": q})
    print(f"답변: {result['answer']}")
````
